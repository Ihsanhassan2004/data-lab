```python
a=int(input("enter the first number"))
b=int(input("enter the second number"))
print("ARITHEMATIC OPERATION :")
print("Addition ",a+b)
print("Subtraction ",a-b)
print("Multiplication ",a*b)
if b==0:
    print("not divisible by zero")
else:
    print("Division ",a/b)
print("Exponation ",a**b)
```

```python
a=input("enter a string")
b=input("enter another string")
c=a+" "+b
print("STRING OPERATIONS ")
print("First string is ",a)
print("concatenation gives",c)
print("Repitation",a*3)
print("indexing",a[0])
print("silicing",a[1:3])
print("Length of a",len(a))
print("replace a to b",b.replace("a","b"))
print("String formatting: hello {}".format(a))
print("capitalize",a.capitalize())
print("Upper case :",a.upper())
print("Justify :",a.rjust(10))
print("To center :",a.center(5))
text="    python "
print("Text stripping :",text.strip())
```
